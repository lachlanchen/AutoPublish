```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import os

def setup_driver():
    chrome_options = Options()
    chrome_options.add_experimental_option("debuggerAddress", "localhost:5005")
    
    # Path to ChromeDriver executable
    chromedriver_path = '/usr/local/bin/chromedriver'  # Adjust this path as necessary
    service = Service(executable_path=chromedriver_path)
    
    # Connect to the existing Chrome session
    driver = webdriver.Chrome(options=chrome_options)
    return driver

def capture_element_screenshot(driver, url, element_selector, output_filename):
    driver.get(url)
    element = driver.find_element(By.CSS_SELECTOR, element_selector)
    
    # Ensure the element is found
    if element:
        element.screenshot(output_filename)
        print(f'Screenshot saved as {output_filename}')
    else:
        print('Element not found')

if __name__ == '__main__':
    driver = setup_driver()
    url = 'https://member.bilibili.com/platform/upload/video/frame'  # Replace with the URL of your choice
    element_selector = '.geetest_holder'  # Replace with your specific selector
    os.makedirs("bilibili_test", exist_ok=True)
    output_filename = os.path.expanduser('bilibili_test/element_screenshot.png')
    
    capture_element_screenshot(driver, url, element_selector, output_filename)
    # driver.quit()
```
